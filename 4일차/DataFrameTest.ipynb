```python
# 피벗테이블 - 데이터 열 중 2개의 열을 추출해 행과 열 구조 재편성
# 2개의 열 관계
#pivot() / pivot_table()
```

```python
data = {
    '도시':['서울', '서울', '서울', '부산', '부산', '부산', '인천', '인천'],
    '연도':['2015', '2010','2005','2015','2010','2005','2015','2010'],
    '인구':[99393938, 99393938,  99393938, 99393938, 99393938, 99393938, 99393938, 99393938],
    '지역':['수도권', '수도권', '수도권', '경상권', '경상권', '경상권', '수도권', '수도권']
}

df6 = pd.DataFrame(data)
df6
```

```python
df6.pivot("도시", "연도", "인구")
```

```python
df6.pivot_table('인구', '도시', '연도', margins_name='평균', margins=True)
```

```python
df6.pivot_table('인구', '도시', '연도', aggfunc=sum, margins_name='합계', margins=True)
```
